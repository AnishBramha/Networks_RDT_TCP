```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('tahoe.csv', header = None, names=['RTT', 'CWND'])
cwnd = df['CWND'].tolist()
rtt = df['RTT'].tolist()

cwnd = [int(x) for x in cwnd[1:]]
rtt = [int(x) for x in rtt[1:]]
```

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10, 5))
plt.plot(rtt, cwnd, marker = 'o', linestyle = '-', color = 'red')
plt.title("TCP Congestion Control (Tahoe)")
plt.xlabel("Transmission Rounds")
plt.ylabel("Congestion Window (cwnd)")
plt.yticks(list(range(1, max(cwnd) + 2)))
plt.grid(True)
plt.tight_layout()
plt.show()
```
